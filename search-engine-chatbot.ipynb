```python
import os
import requests
from langchain_tavily import TavilySearch

# Set Tavily and OpenRouter API Keys
TAVILY_API_KEY = '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'
OPENROUTER_API_KEY = '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'  # Replace with your OpenRouter API key

# Set Tavily API Key
os.environ["TAVILY_API_KEY"] = TAVILY_API_KEY

def openrouter_chat_query(query):
    """Perform query to OpenRouter API (for Qwen model)"""
    base_url = "https://openrouter.ai/api/v1"
    headers = {
        "Authorization": f"Bearer {OPENROUTER_API_KEY}",
        "Content-Type": "application/json",
    }

    data = {
        "model": "qwen/qwen3-0.6b-04-28:free",  # Ensure this model name matches the one you want to use
        "messages": [{"role": "user", "content": query}]
    }

    try:
        response = requests.post(f"{base_url}/chat/completions", headers=headers, json=data)

        # Check if response is successful
        if response.status_code == 200:
            return response.json()['choices'][0]['message']['content']
        else:
            print(f"Error with OpenRouter: {response.status_code}, {response.text}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"Request Error: {e}")
        return None

class QwenChatbot:
    def __init__(self):
        # Initialize Tavily Search Tool
        self.tavily_tool = TavilySearch(
            max_results=5,
            topic="general",
        )
        self.GENERAL_PROMPT_TEMPLATE = """
ROLE:
You are an advanced AI assistant designed to provide comprehensive, accurate, and helpful responses .

WEB SEARCH CONTEXT:
{web_context}

USER QUERY:
{user_query}
"""
    def search_web(self, query):
        try:
            search_results = self.tavily_tool.invoke(query)
            
            if isinstance(search_results, dict) and 'results' in search_results:
                combined_content = "\n".join([result.get('content', 'No content available') for result in search_results['results']])
                return combined_content
            return "No relevant information found."
        
        except Exception as e:
            print(f"Error during search: {e}")
            return "Error occurred during search."

    def generate_response(self, web_content, query):
        full_prompt = self.GENERAL_PROMPT_TEMPLATE.format(
            web_context=web_content,
            user_query=query
        )
        response = openrouter_chat_query(full_prompt)

        if response:
            return response
        else:
            return "I'm sorry, I couldn't generate a response."

    def chat(self):
        print("Qwen Chatbot: Hello! I'm ready to help you. Type 'exit' to quit.")
        while True:
            user_input = input("You: ")
            if user_input.lower() in ['exit', 'quit', 'bye']:
                print("Qwen Chatbot: Goodbye!")
                break
            try:
                search_results = self.search_web(user_input)
                if search_results == "No relevant information found.":
                    response = "I'm sorry, I couldn't find relevant information at the moment."
                else:
                    response = self.generate_response(search_results, user_input)
                print("Qwen Chatbot:", response)
            
            except Exception as e:
                print(f"An error occurred: {e}")
if __name__ == "__main__":
    chatbot = QwenChatbot()
    chatbot.chat()
```
